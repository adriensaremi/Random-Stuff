```python
import random
import numpy as np

import pandas as pd


teams_1 = [['psg','france'],['bayern munich','germany'],['man city','england'],['juventus','italy'],
           ['liverpool','england'],['barcelona','spain'],['leipzig','germany'],['valencia','spain']]
    
teams_2 = [['real madrid','spain'],['tottenham','england'],['atalanta','italy'],['atletico','spain'],
           ['napoli','italy'],['dortmund','germany'],['lyon','france'],['chelsea','england']]
```

```python
n_trials = 10000

matches=[]
for i in range(n_trials):
    list1 = [item for item in range(8)]
    list2 = random.sample([item for item in range(8)],8)
    control = 0
    for j in range(8):
        if list1[j] == list2[j] or teams_1[list1[j]][1] == teams_2[list2[j]][1]:            
            control = 1
            break
    if control == 0:
        matches.append([list1,list2])

        
matches_array = np.array(matches)
len(matches_array)
```

```python
counting = [[0 for x in range(8)] for y in range(8)] 

for n in range(len(matches_array)):
    for k in range(8):
        counting[ matches_array[n,0,k] ][ matches_array[n,1,k] ] += 1
        
counting_rescaled = [  [round(100*j/len(matches_array),2) for j in i] for i in counting] 

df = pd.DataFrame(counting_rescaled,
                   columns =[row[0] for row in teams_2],
                   index = [row[0] for row in teams_1])

df
```
